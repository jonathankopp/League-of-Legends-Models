```python
# Data cleaning (getting rid of dumb attrs and null rows)
# df01 = pd.read_csv('2014.csv')
# df02 = pd.read_csv('2015.csv')
# df03 = pd.read_csv('2016.csv')
df0 = pd.read_csv('2017.csv')
df1 = pd.read_csv('2018.csv')
df2 = pd.read_csv('2019.csv')
df3 = pd.read_csv('2020.csv')

# print(list(set(list(df0['vspm']))))



df = pd.concat([df0,df1,df2,df3])
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(list(set(list(df['league']))))


# df         = pd.DataFrame(data = df,columns = df.columns)
df         = pd.DataFrame(data = df.loc[(df['league'] == 'LCS') | (df['league'] == 'LEC') | (df['league'] == 'LPL')],columns = df.columns)
playerData   = pd.DataFrame(data = df[df['datacompleteness']=='complete'],columns = df.columns)

# seperating team and player data, use teamdata at a different point
# teamData   = pd.DataFrame(data = df[df['position']=='team'], columns = df.columns)
# playerData = pd.DataFrame(data = df[df['position']!='team'], columns = df.columns)



teams   = list(set(list(playerData['team'])))
players = list(set(list(playerData['player'])))
# print(len(playerData))
playerData = playerData[playerData['gamelength'] > 1122]
playerData = playerData[playerData['gamelength'] < 3000]
# print(len(playerData))
playerData['gametime'] = playerData['gamelength']/60
# print(list(set(list(playerData[playerData['position']=='team']['gamelength']))))
playerData.drop(['ban4',
                 'ban5',
                 'gameid',
                 'url',
                 'split',
                 'datacompleteness',
                 'year',
                 'game',
                 'playoffs',
                 'patch',
                 'playerid',
                 'champion',
                 'ban1',
                 'ban2',
                 'ban3',
                 'vspm',
                 'visionscore',
                 'firstbloodassist',
                 'team_kpm',
                 'dragons_type_unknown',
                 'teamkills',
                 'teamdeaths',
                 'doublekills',
                 'triplekills',
                 'quadrakills',
                 'pentakills',
                 'earnedgoldshare',
                 'minionkills',
                 'monsterkills',
                 'goldat10',
                 'opp_goldat10',
                 'goldat15',
                 'opp_goldat15',
                 'opp_csat10',
                 'opp_csat15',
                 'gspd',
                 'firstblood',
                 'total_cs',
                 'earnedgold',
                 'totalgold',
                 'goldspent',
                 'wardsplaced',
                 'wardskilled',
                 'ckpm',
                 'opp_xpat10',
                 'opp_xpat15',
                 'opp_goldat10',
                 'opp_goldat15',
                 'damagetochampions',
                 'controlwardsbought',
                 'damagetakenperminute',
                 'damagemitigatedperminute',
          'league'],axis=1,inplace = True)
for col in playerData.columns:
    print(col)
```

```python
# contextualizing features/ feature engineering/ plz work daddy
playerData.side.replace(('Blue', 'Red'), (0, 1), inplace=True)
playerData['deaths'].replace((0),(1),inplace=True)
playerData['kd'] = round(playerData['kills']/playerData['deaths'],4)


playerData['apm'] = round(playerData['assists']/playerData['gametime'],2)
playerData['mkoj_pm'] = round(playerData['monsterkillsownjungle']/playerData['gametime'],2)
playerData['mkej_pm'] = round(playerData['monsterkillsenemyjungle']/playerData['gametime'],2)

                 
# playerData.dropna(axis = 0, subset = playerData.columns, how='any', inplace=True)
playerData.drop(['kills',
                 'deaths',
                 'assists',
                 'monsterkillsownjungle',
                 'monsterkillsenemyjungle',
                 'gametime',
                 'gamelength'
                ],axis=1,inplace=True)
playerData.fillna(0,inplace=True)
```

```python
# # this gets rid of uneeded attrs (feature selection)
for col in list(playerData.columns):
    if('opp_' in col and 'cs' not in col and 'xp' not in col):
        playerData.drop([col],axis=1,inplace=True)
        continue
    elif('opp_'+col in list(playerData.columns) and 'cs' not in col and 'xp' not in col):
        playerData[col+"diff"] = playerData[col] - playerData['opp_'+col]
        playerData.drop([col],axis=1,inplace=True)

# # Using minmaxscaler to make values inbetween 0 and 1
# mms        = MinMaxScaler(feature_range=(0,1))
# mms        = mms.fit(featSelect)
# scaledData = mms.transform(featSelect)
# print(scaledData)
playerData.head(12)

# Data getting from team row
# list(playerData.columns[8:18]) + list(playerData.columns[38:])
playerData.columns
# len(playerData.columns)
```

```python
import pandas as pd
import json
import numpy as np

df      = pd.read_csv('LCS_Data/preSweep.csv',index_col='date',parse_dates = True)
prepTeamDat = []
prepPlayerDat = []
rpm = []
players = dict()
playerC = dict()
sPlayer = dict()
tPlayer = dict()
print(df.shape)
# print(df.columns,len(df.columns))

# test = pd.DataFrame.from_dict(data=leg,orient='index',columns=['nip','size'])

# create player data array, and then in 'Real Time' use those to create rows for the model

# print(df.head(12))
gameAmntAdd = 0

counter  = 1
team     = 0
exp      = 0
gameAmnt = 6
realCount = 1
teamR   = np.array([0 for i in range(17)])
teamB   = np.array([0 for i in range(17)])
temp   = {
    'top' : [],
    'jng' : [],
    'mid' : [],
    'bot' : [],
    'sup' : []
}
skip = False
sCount = 0

# print(list(df[df['position']=='team']['opp_dragons']).count(-99),len(list(df[df['position']=='team']['dragons'])))
# print(list(set(list(df[df['position']!='team']['csat10']))))
# change to a type of moving avg later on for emphasis on new data
for i,row in df.iterrows():
    if(skip and sCount < 2):
        if(row['position'] != 'team'):
            print('fuck')
            print(counter)
            break
        realCount += 1
        sCount+=1
        continue
    elif(skip and sCount >= 2):
        skip = False
        sCount = 0
    if(row['position']=='team'):
        print('heredafuq')
        break
    player = row['player']
    playI  = counter%10 if counter%10 !=0 else 10
    if player not in players.keys():
        tPlayer[player] = np.array([0 for i in range(17)])
        players[player] = np.array([0 for i in range(39)])
        sPlayer[player] = np.array([0 for i in range(22)])
        playerC[player] = 1
    if(counter%5 != 0):
        if(team == 0):
            teamR = np.add(teamR,tPlayer[player])/2.0
        else:
            teamB = np.add(teamB,tPlayer[player])/2.0
        temp[row['position']] = np.append(temp[row['position']], sPlayer[player])

    elif(counter%5 == 0):
        if(team == 0):
            teamR = np.add(teamR,tPlayer[player])/2.0
        else:
            teamB = np.add(teamB,tPlayer[player])/2.0
        temp[row['position']] = np.append(temp[row['position']], sPlayer[player])

        if(counter%10 == 0):
            # every player has >= gameAmnt games]
            if(exp >= 8):
                side = 'Red' if row['side'] == 0 else 'Blue'
                if(row['result']==1):
                    prepTeamDat.append(np.append([gameAmntAdd],np.append(np.concatenate((teamR, teamB), axis=None),[side])))
                    prepPlayerDat.append({
                        gameAmntAdd : {
                            'top': np.append(temp['top'],side),
                            'jng': np.append(temp['jng'],side),
                            'mid': np.append(temp['mid'],side),
                            'bot': np.append(temp['bot'],side),
                            'sup': np.append(temp['sup'],side)
                        },
                    })
                    rpm.append(np.append(gameAmntAdd,np.append('top', np.append(temp['top'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('jng', np.append(temp['jng'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('mid', np.append(temp['mid'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('bot', np.append(temp['bot'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('sup', np.append(temp['sup'],side))))

                else:
                    side = 'Red' if row['side'] == 1 else 'Blue'
                    prepTeamDat.append(np.append([gameAmntAdd],np.append(np.concatenate((teamR, teamB), axis=None),[side])))
                    prepPlayerDat.append({
                        gameAmntAdd : {
                            'top': np.append(temp['top'],side),
                            'jng': np.append(temp['jng'],side),
                            'mid': np.append(temp['mid'],side),
                            'bot': np.append(temp['bot'],side),
                            'sup': np.append(temp['sup'],side)
                        },
                    })
                    rpm.append(np.append(gameAmntAdd,np.append('top', np.append(temp['top'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('jng', np.append(temp['jng'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('mid', np.append(temp['mid'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('bot', np.append(temp['bot'],side))))
                    rpm.append(np.append(gameAmntAdd,np.append('sup', np.append(temp['sup'],side))))
                gameAmntAdd += 1  
                # print(len(prepTeamDat[0][0]))
                # break
            exp = 0
            skip=True
            temp   = {
                'top' : [],
                'jng' : [],
                'mid' : [],
                'bot' : [],
                'sup' : []
            }
            
    # logic for if the amount of games played is over 10(or whatever we decide) create the rows
    if(playerC[player] >= gameAmnt):
        exp += 1
    # logic for adding the current data to the playerdict
    location = 0
    if(team==0):
        location=10
    else:
        location=11
    location -= playI

    teamData = df.iloc[realCount+(location)]
    if(teamData['position']!='team'):
        print(realCount,'bfuckitch',row['player'])
        break
    # print(teamData['position'],teamData['team'],playI,realCount+(location),team)
    teamDataList1 = list(teamData[7:17])
    teamDataList2 = list(teamData[37:])
    comp = list(row[5:7])+teamDataList1+list(row[17:37])+teamDataList2
    
    fullTeam  = teamDataList1 + teamDataList2
    # add add then average it
    tPlayer[player] = np.add(np.array(fullTeam),tPlayer[player])/2.0
    players[player] = np.add(np.array(comp),players[player])/2.0
    sPlayer[player] = np.add(np.array(list(row[5:7])+list(row[17:37])),sPlayer[player])/2.0
    playerC[player] += 1
    if(counter%5==0):
        team = 1 if team == 0 else 0
    counter += 1
    realCount +=1
# print(players['Rascal'])
```
